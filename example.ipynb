```python
coords = app.data["textblocks"][0][3]
```

```python
import pytesseract
from pytesseract import Output
import cv2

d = pytesseract.image_to_data(app.full_img, output_type=Output.DICT)
```

```python
import PIL
import numpy
pil_image = app.full_img.crop(coords).convert('RGB') 
open_cv_image = numpy.array(pil_image) 
# Convert RGB to BGR 
img = open_cv_image[:, :, ::-1].copy() 
```

```python
n_boxes = len(d['level'])
for i in range(n_boxes):
    (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
    cv2.rectangle(img, (x, y), (x + w, y + h), 50, 2)

cv2.imshow('img', img)
cv2.waitKey(0)
```
